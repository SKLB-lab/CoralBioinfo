```r
options(repr.plot.width=40,repr.plot.height=15)
data = read.table("Acropora_muricata.miClus",header=FALSE,row.names=1)
data=data[!rownames(data) %in% c('159855nm265','159855nm187'),]
dists = dist(data,method = "euclidean")
hc = hclust(dists, method = "ave")
plot(hc)
```
